```python
import seaborn as sns
tips = sns.load_dataset("tips")
df = tips.copy()
```

```python
df.head()
```

```python
sns.catplot(y="total_bill", kind="violin",data=df);
```

```python
#cross-correlation
```

```python
sns.catplot(x ="day", y="total_bill", kind = "violin", data=df);
```

```python
sns.catplot(x ="day", y="total_bill", kind = "violin", hue="sex", data=df);
```
