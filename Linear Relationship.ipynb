# Linear Relationship

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
df=tips.copy()
```

```python
df.head()
```

```python
sns.lmplot(x = "total_bill", y = "tip", data = df);
```

```python
sns.lmplot(x = "total_bill", y = "tip", hue = "smoker",  data = df);
```

```python
sns.lmplot(x = "total_bill", y = "tip", col="time", hue = "smoker",  data = df);
```

```python
sns.lmplot(x = "total_bill", y = "tip", col="time", row="sex", hue = "smoker",  data = df);
```

```python
sns.lmplot(x = "total_bill", y = "tip", col="time", row="sex", hue = "smoker", palette="Spectral",  data = df);
```
